# Decision Tree Classifier

```python
# importing Libraries


import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
```

```python
iris=datasets.load_iris()
```

```python
# preparind data into dataframes and column names
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df.head(5)
```

```python
print(iris.target)
```

```python
from sklearn.model_selection import train_test_split
```

# Intializing classifier

```python
clf=DecisionTreeClassifier()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(iris.data,iris.target,random_state=1)
```

```python
clf.fit(X_train,Y_train)

print("Decision Tree Trained...")
```

```python
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image 
```

```python
dot_data = export_graphviz(clf, out_file=None,
                          feature_names=iris.feature_names,
                          class_names=iris.target_names)
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_pdf("iris.pdf")
Image(graph.create_png())
```

# Now let us predict the data

```python
# predicting on training data and predicting on testing data
y_train_pred=clf.predict(X_train)
y_test_pred=clf.predict(X_test)
```

# importing confusion matrix and printing it

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(Y_train,y_train_pred)
```

```python
confusion_matrix(Y_test,y_test_pred)
```

from above we can see that our classifier is working well on both training and testing data

```python
from sklearn.metrics import classification_report
print("CLASSIFICATION REPORT:")
print(classification_report(Y_test,y_test_pred))
```
